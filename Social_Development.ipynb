```python
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
import numpy as np 

# method to fromat the data for a line chart
# I can still write her 
def top(df_top):
    #create a empty datafram with said colmns
    df_formated = pd.DataFrame(columns = ["Country Name","year","data"])

    list_name = []

    #append the data for every year of every country
    for y in df_top.columns[4:24]:
        for z in df_top[y]:
            df_formated = df_formated.append({"year":y,"data":z},ignore_index=True)

    #get all of the Country Names in a list
    for x in range(0,20):
        temp_list = df_top["Country Name"].values
        for x in temp_list:
            list_name.append(x)

    #add the list of names to column Country Name
    df_formated["Country Name"] = list_name

    #sort the dataframe by both Country Name and year
    df_formated = df_formated.sort_values(["Country Name","year"])

    return df_formated

#read the csvs
df = pd.read_csv("data.csv")
df_income = pd.read_csv("income.csv")

year_list = []

#change the column names to the values in row 4
df.columns = df.iloc[3]

#drop the first 4 rows
df = df.drop(df.index[0:4])

# years were removing
for x in range(1960,2001):

    year_list.append(x)

#drop all of the columns with no data/little data
for x in year_list:

    df = df.drop(columns=[x])

#drop any rows with any None values
df = df.dropna()

#limit the columns to these
df_income = df_income[["Country Code","IncomeGroup"]]

# merge the two data frames on Country Code
df_merge = df.merge(df_income,how="inner",on="Country Code")

#df_merge = df_merge.dropna()

# all of the column names are numbers so change them all to strings
for x in range(2001,2021):
    df_merge = df_merge.rename(columns={x : str(x)})

#save the cleaned data to a csv
df_merge.to_csv("cleaned_data.csv")

#drop any rows with none values
df_merge = df_merge.dropna()

#create new column to show the change in the perecent from 2001 to 2020
df_merge["Rate of Change"] = df_merge["2020"] - df_merge["2001"]

df_merge
```

```python
#create new dataframes per Incomegroup and then sort them by rate of change
df_high = df_merge[df_merge["IncomeGroup"] == "High income"]
df_high = df_high.sort_values("Rate of Change",ascending=False)
df_low = df_merge[df_merge["IncomeGroup"] == "Low income"]
df_low = df_low.sort_values("Rate of Change",ascending=False)
df_low_mid = df_merge[df_merge["IncomeGroup"] == "Lower middle income"]
df_low_mid = df_low_mid.sort_values("Rate of Change",ascending=False)
df_up_mid = df_merge[df_merge["IncomeGroup"] == "Upper middle income"]
df_up_mid = df_up_mid.sort_values("Rate of Change",ascending=False)

#create the instance of the app
app = dash.Dash(__name__)

#html layout
app.layout = html.Div([
    html.P("Years:"), #create a <p> tag with text
    dcc.Dropdown( #create a dropdown element
        id='years', #id for the dropdown element
        options=[{'value': x, 'label': x} #value is the value passed @app.callback and label is the display
                 for x in ['2001',"2020"]], # list of the options
        clearable=True # this allows you to clear the dropdown element
    ),
    html.P("Income:"),
    dcc.Dropdown( 
        id='income',
        options=[{'value': x, 'label': x} 
                 for x in ['High income',"Low income","Lower middle income","Upper middle income"]],
        clearable=False
    ),
    html.P("Top 5:"), 
    dcc.Dropdown(
        id='movers',
        options=[{'value': x, 'label': x} 
                 for x in ['Top 5 Increases',"Top 5 Decreases"]],
        clearable=True
    ),
    dcc.Graph(id="chart"), #create a graph element to display the plotly figure
])

@app.callback( # this gets called everytime a element is changed (dropdown option changes)
    Output("chart", "figure"), # chart is the graph element and figure is the return from generate_chart
    [Input("years", "value"), # years is the id for one of the dropdown abd the value gets passed to generate_chart
    Input("income", "value"), # income is the id for one of the dropdown abd the value gets passed to generate_chart
    Input("movers", "value")]) # movers is the id for one of the dropdown abd the value gets passed to generate_chart
     
def generate_chart(values,income,movers): # this gets called with the values from the input of @app.callback
    if income == "High income": # checks to see with income dropdown option is picked
        fig = px.bar(df_high,x = "Country Name",y = values,text_auto = ".3s") # create a bar chart with the values from high income dataframe
        if movers == "Top 5 Increases": # checks to see if there is an option picked for top 5 dropdown element
            fig = px.line(top(df_high.head(5)),x = "year", y = "data",color='Country Name') # create a line chart with the data from the top 5 elements in the high income dataframe
        if movers == "Top 5 Decreases": # same but bottom 5
            fig = px.line(top(df_high.tail(5)),x = "year", y = "data",color='Country Name')

    elif income == "Low income":
        fig = px.bar(df_low,x = "Country Name",y = values,text_auto = ".3s")
        if movers == "Top 5 Increases":
            fig = px.line(top(df_low.head(5)),x = "year", y = "data",color='Country Name')
        if movers == "Top 5 Decreases":
            fig = px.line(top(df_low.tail(5)),x = "year", y = "data",color='Country Name')

    elif income == "Lower middle income":
        fig = px.bar(df_low_mid,x = "Country Name",y = values,text_auto = ".3s")
        if movers == "Top 5 Increases":
            fig = px.line(top(df_low_mid.head(5)),x = "year", y = "data",color='Country Name')
        if movers == "Top 5 Decreases":
            fig = px.line(top(df_low_mid.tail(5)),x = "year", y = "data",color='Country Name')

    elif income == "Upper middle income":
        fig = px.bar(df_up_mid,x = "Country Name",y = values,text_auto = ".3s")
        if movers == "Top 5 Increases":
            fig = px.line(top(df_up_mid.head(5)),x = "year", y = "data",color='Country Name')
        if movers == "Top 5 Decreases":
            fig = px.line(top(df_up_mid.tail(5)),x = "year", y = "data",color='Country Name')

    return fig

#start the app
app.run_server()
```
